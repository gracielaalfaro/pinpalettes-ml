```python
def plot_palette(hex_colors, proportions):
    fig, ax = plt.subplots(figsize=(8,2))
    x = 0
    for hex_color, prop in zip(hex_colors, proportions):
        ax.barh(0, width=prop, left=x, color=hex_color)
        x += prop
    ax.axis("off")
    plt.show()

plot_palette(hexs, props)

```
